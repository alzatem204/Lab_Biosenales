```python
# importaci칩n de librer칤as
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt
from scipy.signal import welch, firwin, filtfilt
```

***Creaci칩n de la se침al***

```python
# creaci칩n de x_e, y_e, w_e
F0 = 0.01 # frecuencia fundamental
Fs = 25 #frecuencia de muestreo
alfa = 0.8 # atenuaci칩n
D = 15 # retrado induciodo en la adquisici칩n
var = 0.01 # varianza del ruido
media = 0 # media del ruido
n = np.arange(0,1001,1/Fs) # tiempo
x_e = np.sin(2*np.pi*F0*(n+15/Fs)) #potencial con con retraso
w = np.random.normal(media,np.sqrt(var),len(n)) # ruido gausiano
y_e = alfa*x_e + w # se침al adquirida con ruido y retrado.

plt.plot(n,y_e)
plt.ylabel("Amplitud")
plt.xlabel("Tiempo (s)")
```

***Dise침o de un filtro FIR pasa bajas para la se침al 洧녽洧(洧녵), con una frecuencia de corte
de 1.25 muestras/s y orden 30.***

```python
order = 30 # orden calculado segun requerimientos
nyquist = Fs/2 # frecuencia de nyquist
wn = 1.25/nyquist #frecuencia de corte normalizada respecto nyquist
lowpass = signal.firwin(order+1,wn, pass_zero = 'lowpass',window='hamming') #filtro
lowpass_32 = signal.firwin(33+1,wn, pass_zero = 'lowpass',window='hamming') #filtro
y_f_30 = signal.filtfilt(b=lowpass,a=1,x=y_e) # aplicaci칩n del filtro
y_f_32 = signal.filtfilt(b=lowpass_32,a=1,x=y_e) # aplicaci칩n del filtro

plt.plot(n,y_f_30)
plt.plot(n,y_f_32)
plt.ylabel("Amplitud")
plt.xlabel("Tiempo (s)")
```

***쯈u칠 cambios hay en los argumentos de la funci칩n para implementar un filtro
pasa banda y pasa altas? Pruebe con una frecuencia de corte adicional de 7.5
muestras/s.***

```python
# Frecuencia de muestreo de la se침al original
fs = Fs

# Calcular el periodograma de Welch de la se침al original
frecuencias_orig, densidad_espectral_orig = welch(y_e, fs, nperseg=1024)

# Dise침ar un filtro FIR pasa altas con frecuencia de corte adicional de 7.5 muestras/s
frecuencia_corte_pa = 7.5 / (fs / 2)
orden = 101
coeficientes_filtro_pa = firwin(orden+1, frecuencia_corte_pa, pass_zero=True, window='hamming')


# Filtrar la se침al utilizando filtfilt para mantener la fase
se침al_filtrada = filtfilt(coeficientes_filtro_pa, 1, y_e)

# Calcular el periodograma de Welch de la se침al filtrada
frecuencias_filtrada, densidad_espectral_filtrada = welch(se침al_filtrada, fs, nperseg=1024)

# Definir el tiempo para la se침al original
tiempo = np.arange(0, len(y_e) / fs, 1 / fs)

# Graficar un segmento de las se침ales antes y despu칠s de filtrar
plt.figure(figsize=(12, 6))
plt.plot(tiempo, y_e, label='Se침al Original', color='blue')
plt.plot(tiempo, se침al_filtrada, label='Se침al Filtrada', color='red')
plt.title('Comparaci칩n de la Se침al Original y Filtrada')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()

# Graficar el periodograma de Welch de la se침al original y filtrada
plt.figure(figsize=(12, 6))
plt.semilogy(frecuencias_orig, densidad_espectral_orig, label='Original', color='blue')
plt.semilogy(frecuencias_filtrada, densidad_espectral_filtrada, label='Filtrada', color='red')
plt.title('Periodograma de Welch de la Se침al Original y Filtrada')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Densidad Espectral de Potencia')
plt.legend()
plt.grid(True)
plt.show()
```

An치lisis: Al ajustar los argumentos de la funci칩n para implementar un filtro pasa banda y pasa altas con una frecuencia de corte adicional de 7.5 muestras/s, se logra adaptar el comportamiento del filtro a las necesidades espec칤ficas. La imagen detalla la respuesta en frecuencia de un filtro pasa banda con un orden de 30 y frecuencias de corte de 1.25 y 7.5 muestras/s, presentando la magnitud, la magnitud en decibeles y la fase. Esta representaci칩n revela que el filtro permite el paso de frecuencias dentro de la banda de paso mientras aten칰a las frecuencias fuera de esta banda, con una atenuaci칩n mayor a medida que la frecuencia se aleja de la banda de paso. El orden del filtro de 30 proporciona una banda de paso estrecha y una notable atenuaci칩n fuera de esta banda, esto es demasiado importante, ya que podemos concluir que el filtro pasa banda con un orden de 30 y frecuencias de corte de 1.25 y 7.5 muestras/s normalizadas es ideal para aplicaciones que requieren el paso de frecuencias dentro de un rango espec칤fico y la atenuaci칩n de las frecuencias fuera de ese rango, gracias a su banda de paso estrecha y su significativa atenuaci칩n fuera de esta banda.

***Resultado esperado dise침o de 3 filtros FIR (pasa bajas, pasa altas y pasa banda)***

```python
import matplotlib.pyplot as plt
from scipy.signal import firwin, filtfilt

# Definir orden del filtro y frecuencia de Nyquist
order = 30
nyquist = Fs / 2

# 1. Filtro FIR pasa bajas
wn_lowpass = 1.25 / nyquist  # Frecuencia de corte normalizada para pasa bajas
filtro_pasa_bajas = firwin(order + 1, wn_lowpass, pass_zero='lowpass')
y_f_lowpass = filtfilt(filtro_pasa_bajas, 1, y_e)  # Aplicar filtro pasa bajas

# 2. Filtro FIR pasa altas
wn_highpass = 7.5 / nyquist  # Frecuencia de corte normalizada para pasa altas
filtro_pasa_altas = firwin(order + 1, wn_highpass, pass_zero='highpass')
y_f_highpass = filtfilt(filtro_pasa_altas, 1, y_e)  # Aplicar filtro pasa altas

# 3. Filtro FIR pasa banda
wn_bandpass = [1.25 / nyquist, 7.5 / nyquist]  # Frecuencias de corte normalizadas para pasa banda
filtro_pasa_banda = firwin(order + 1, wn_bandpass, pass_zero='bandpass')
y_f_bandpass = filtfilt(filtro_pasa_banda, 1, y_e)  # Aplicar filtro pasa banda

# Crear subplots para cada filtro
plt.figure(figsize=(12, 8))

# Se침al original
plt.subplot(4, 1, 1)
plt.plot(n, y_e, label='Se침al Original', color='blue')
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.title("Se침al Original")

# Se침al filtrada con filtro pasa bajas
plt.subplot(4, 1, 2)
plt.plot(n, y_f_lowpass, label='Se침al Filtrada - Pasa Bajas', color='orange')
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.title("Filtro FIR Pasa Bajas")

# Se침al filtrada con filtro pasa altas
plt.subplot(4, 1, 3)
plt.plot(n, y_f_highpass, label='Se침al Filtrada - Pasa Altas', color='green')
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.title("Filtro FIR Pasa Altas")

# Se침al filtrada con filtro pasa banda
plt.subplot(4, 1, 4)
plt.plot(n, y_f_bandpass, label='Se침al Filtrada - Pasa Banda', color='red')
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.title("Filtro FIR Pasa Banda")

plt.tight_layout()
plt.show()
```

***Dise침e un filtro IIR en las mismas condiciones de los puntos anteriores. Compare
cambios.***

```python
# Par치metros del sistema
Fs = 25  # Frecuencia de muestreo
nyquist = Fs / 2  # Frecuencia de Nyquist

# Dise침o de los filtros IIR
# Filtro pasa bajas
b_low, a_low = signal.iirfilter(N=3, Wn=1.25/nyquist, btype='low', ftype='butter')
y_iir_low = signal.filtfilt(b_low, a_low, y_e)

# Filtro pasa altas
b_high, a_high = signal.iirfilter(N=3, Wn=7.5/nyquist, btype='high', ftype='butter')
y_iir_high = signal.filtfilt(b_high, a_high, y_e)

# Filtro pasa banda
b_band, a_band = signal.iirfilter(N=3, Wn=[1.25/nyquist, 7.5/nyquist], btype='band', ftype='butter')
y_iir_band = signal.filtfilt(b_band, a_band, y_e)

# Gr치ficos
plt.figure(figsize=(12, 8))

# Se침al original
plt.subplot(4, 1, 1)
plt.plot(n, y_e, label='Se침al Original', color='black')
plt.title('Se침al Original')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')

# Se침ales filtradas con filtros IIR
plt.subplot(4, 1, 2)
plt.plot(n, y_iir_low, label='Filtro IIR Pasa Bajas', color='blue')
plt.title('Filtro IIR Pasa Bajas')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')

plt.subplot(4, 1, 3)
plt.plot(n, y_iir_high, label='Filtro IIR Pasa Altas', color='red')
plt.title('Filtro IIR Pasa Altas')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')

plt.subplot(4, 1, 4)
plt.plot(n, y_iir_band, label='Filtro IIR Pasa Banda', color='green')
plt.title('Filtro IIR Pasa Banda')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')

plt.tight_layout()
plt.show()
```
