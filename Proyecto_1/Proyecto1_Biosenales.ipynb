```python
def hampelfilter(wk, K, FilterParms):
    """
    Procedure to implement the Hampel filter
    """
    #
    Thresh = FilterParms[0]
    ctr = wk#[K]
    ref = np.median(wk)

    AbsDev = abs(wk - ref)
    MAD = 1.4826 * np.median(AbsDev)
    TestVal = abs(ctr - ref)
    #if TestVal > Thresh * MAD:
    yk = ctr
    numero = np.sum((TestVal > Thresh * MAD)==True)

    yk[TestVal > Thresh * MAD] = ref


    return yk
```

```python
# Esta es la funci贸n que calcula la entrop铆a de permutaci贸n
def permutation_entropy(signal, dim=4, t=1):

  # Modulo numpy Funci贸n de ventana deslizante
  sorted_matrix = sliding_window_view(signal, dim, axis=0).T[:, ::t]
  # Patrones ordinales y ranking
  index_matrix = np.argsort(sorted_matrix, axis=0)
  #Busca lo arreglos diferentes y las veces que se repite
  _ , counts = np.unique(index_matrix, axis=1, return_counts=True)
  #numero de columnas de la matriz
  n = len(signal)- (dim-1) * t
  # frecuencia relativa de ocurrencia de cada permutaci贸n 煤nica en la matriz
  relative_freq = counts / n
  # Entrop铆a de Permutaci贸n a partir de la entropia de shannon
  PE= np.sum(- relative_freq * np.log2(relative_freq))
  # Se retorna el valor de la entropia y la matriz de valores organizados
  return PE
```

# Pruebas Estadisticas

En el presente an谩lisis, se inici贸 con la evaluaci贸n de la normalidad de los datos mediante la prueba de Shapiro-Wilk, con el fin de determinar si las variables de inter茅s segu铆an una distribuci贸n normal, este paso es crucial en el an谩lisis estad铆stico, ya que la normalidad de los datos influye directamente en la selecci贸n de las pruebas adecuadas para la comparaci贸n entre grupos.

```python
import pandas as pd
import scipy.stats as stats

# Funci贸n para evaluar la normalidad de los datos
def evaluar_normalidad(data, variable):
    stat, p_value = stats.shapiro(data[variable])

    print(f"Resultados de la Prueba de Shapiro-Wilk para {variable}:")
    print(f"Estad铆stico: {format(stat, '.50f')}")
    print(f"Valor p: {format(p_value, '.50f')}")

    # Planteamiento de hip贸tesis
    if p_value > 0.05:
        print(f"Para la variable {variable}: No se rechaza la hip贸tesis nula (H). Los datos parecen seguir una distribuci贸n normal.")
        print("Prueba recomendada: Param茅trica\n")
        return "Param茅trica"
    else:
        print(f"Para la variable {variable}: Se rechaza la hip贸tesis nula (H). Los datos no siguen una distribuci贸n normal.")
        print("Prueba recomendada: No param茅trica\n")
        return "No param茅trica"

# Evaluar normalidad en todas las variables de inter茅s
def evaluar_todas_las_variables(df, variables):
    resultados = {}
    for variable in variables:
        if variable in df.columns:
            tipo_prueba = evaluar_normalidad(df, variable)
            resultados[variable] = tipo_prueba
        else:
            print(f"La variable {variable} no se encuentra en el DataFrame.")

    return resultados

# Variables de inter茅s
variables = ['sd_intervalo_izq', 'cv_intervalo_izq', 'ent_intervalo_izq',
             'sd_intervalo_der', 'cv_intervalo_der', 'ent_intervalo_der',
             'sd_balanceo_izq', 'cv_balanceo_izq', 'ent_balanceo_izq',
             'sd_balanceo_der', 'cv_balanceo_der', 'ent_balanceo_der']

# Aplicar la evaluaci贸n de normalidad a todas las variables
resultados_normalidad = evaluar_todas_las_variables(df_analisis, variables)

# Mostrar los resultados resumidos
#for variable, tipo_prueba in resultados_normalidad.items():
    #print(f"Variable: {variable} | Prueba recomendada: {tipo_prueba}")
```

En el contexto de este estudio, los resultados de la prueba de normalidad indicaron que la mayor铆a de las variables no cumpl铆an con el supuesto de normalidad, lo que llev贸 a la decisi贸n de descartar las pruebas param茅tricas, que requieren esta condici贸n.

A pesar de que un an谩lisis adicional de homocedasticidad es decir, la igualdad de varianzas entre grupos es com煤n en este tipo de an谩lisis, se opt贸 por no realizar esta evaluaci贸n. Esta decisi贸n se fundament贸 en que, dado que los datos ya mostraban una falta de normalidad, la evaluaci贸n de la homocedasticidad se volv铆a menos relevante en este contexto.

Las pruebas no param茅tricas, como la prueba de Kruskal-Wallis,que es adecuada para comparar tres o m谩s grupos independientes cuando los datos no cumplen con los supuestos de normalidad, no requieren que las varianzas sean homog茅neas, lo que las convierte en una opci贸n m谩s apropiada para el an谩lisis realizado.

```python
import pandas as pd
import scipy.stats as stats

# Funci贸n para realizar la prueba de Kruskal-Wallis
def kruskal_test(df, variable, group_column):
    grupos = [group[variable].values for name, group in df.groupby(group_column)]
    stat, p_value = stats.kruskal(*grupos)

    print(f"Resultados de la Prueba de Kruskal-Wallis para {variable}:")
    print(f"Estad铆stico: {format(stat, '.50f')}")
    print(f"Valor p: {format(p_value, '.50f')}")

    if p_value < 0.05:
        print("Se rechaza la hip贸tesis nula (H). Hay diferencias significativas entre los grupos.\n")
    else:
        print("No se rechaza la hip贸tesis nula (H). No hay diferencias significativas entre los grupos.\n")

# Variables de inter茅s
variables = ['sd_intervalo_izq', 'cv_intervalo_izq', 'ent_intervalo_izq',
             'sd_intervalo_der', 'cv_intervalo_der', 'ent_intervalo_der',
             'sd_balanceo_izq', 'cv_balanceo_izq', 'ent_balanceo_izq',
             'sd_balanceo_der', 'cv_balanceo_der', 'ent_balanceo_der']

# Aplicar la prueba de Kruskal-Wallis para cada variable
for variable in variables:
    kruskal_test(df_analisis, variable, 'Estado')
```

Este enfoque se alinea con el objetivo de obtener resultados robustos y fiables, asegurando que las conclusiones sobre las diferencias entre los grupos cl铆nicos (como Parkinson, ELA y Huntington) y el grupo control sean v谩lidas, a pesar de la falta de normalidad en los datos. As铆, se prioriz贸 la validez de los resultados sobre la rigidez de los supuestos tradicionales en el an谩lisis param茅trico. De este modo, se logr贸 un an谩lisis m谩s acorde con las caracter铆sticas de los datos disponibles, garantizando la integridad del estudio y la pertinencia de las inferencias realizadas.
