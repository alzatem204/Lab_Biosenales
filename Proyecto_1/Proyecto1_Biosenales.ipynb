**¿Qué es la enfermedad de Parkinson?**
La enfermedad de Parkinson (PD) es un trastorno neurodegenerativo crónico que afecta el sistema nervioso central, particularmente las áreas del cerebro responsables del control del movimiento, es tambien la segunda enfermedad neurodegenerativa más común, después de la enfermedad de Alzheimer, y afecta a millones de personas en todo el mundo. Su principal característica es la pérdida progresiva de neuronas dopaminérgicas en la sustancia negra, una región del cerebro que juega un papel crucial en la regulación del movimiento, esta pérdida de dopamina se traduce en una serie de síntomas motores, que incluyen temblores, rigidez muscular, bradicinesia (movimiento lento) y dificultades en la coordinación y el equilibrio, además de los síntomas motores, los pacientes pueden experimentar una variedad de síntomas no motores, como trastornos del sueño, depresión, ansiedad, problemas de memoria y deterioro cognitivo, estos síntomas no motores a menudo se pasan por alto en el diagnóstico inicial, aunque pueden tener un impacto significativo en la calidad de vida del paciente. La enfermedad de Parkinson es progresiva, lo que significa que los síntomas tienden a empeorar con el tiempo, lo que subraya la importancia de una intervención temprana y un manejo adecuado para mejorar la calidad de vida de quienes la padecen.

**¿Cómo se diagnostica la enfermedad de Parkinson?**
El diagnóstico de la enfermedad de Parkinson es un proceso complejo que involucra una evaluación exhaustiva de los síntomas clínicos y la historia médica del paciente, así como una serie de exámenes neurológicos. En general, no existe una única prueba diagnóstica que pueda confirmar la enfermedad de manera definitiva; más bien, el diagnóstico se basa en la identificación de un conjunto de síntomas motores y no motores. La evaluación clínica es fundamental, ya que un neurólogo experimentado puede observar y evaluar los signos clínicos característicos, como la presencia de temblores en reposo, rigidez y cambios en la marcha, además, se pueden utilizar técnicas de imagen, como la tomografía por emisión de positrones (PET) o la tomografía computarizada por emisión de fotón único (SPECT), para visualizar la pérdida de neuronas dopaminérgicas en el cerebro. Un ejemplo de esto es el DaTscan, que permite evaluar la cantidad de transportadores de dopamina disponibles en el cerebro. También se realizan evaluaciones neuropsicológicas para detectar síntomas cognitivos y no motores que pueden estar presentes. En algunos casos, la respuesta a tratamientos como la levodopa, que se utiliza para aumentar los niveles de dopamina, puede servir como un indicador adicional que respalda el diagnóstico, sin embargo, el proceso de diagnóstico puede ser desafiante, especialmente en las etapas iniciales, debido a la variabilidad en la presentación de los síntomas.

**Retos para el diagnóstico temprano**
El diagnóstico temprano de la enfermedad de Parkinson presenta múltiples desafíos que pueden dificultar una intervención oportuna y eficaz, uno de los principales retos es que los síntomas iniciales suelen ser vagos y pueden confundirse fácilmente con otras condiciones menos graves o con el envejecimiento normal. Por ejemplo, cambios sutiles en la marcha, la postura o el estado de ánimo pueden pasar desapercibidos tanto para los pacientes como para los médicos, lo que retrasa el diagnóstico, además, la falta de biomarcadores específicos que puedan indicar la presencia de la enfermedad en sus primeras etapas complica aún más la situación. Actualmente, la acumulación de proteínas como la alfa-sinucleína, un signo clave de la enfermedad, solo se puede confirmar mediante autopsia, lo que significa que no hay un método de diagnóstico fiable disponible en la práctica clínica. La variabilidad en la presentación de la enfermedad también juega un papel importante; cada paciente puede experimentar diferentes combinaciones de síntomas, y la progresión de la enfermedad puede diferir significativamente entre individuos, esto no solo puede complicar el diagnóstico, sino también influir en las decisiones de tratamiento. A medida que la investigación avanza, hay un creciente interés en el desarrollo de nuevas tecnologías y biomarcadores que podrían facilitar un diagnóstico más temprano y preciso, sin embargo, estos enfoques aún se encuentran en fases de estudio y necesitan validación antes de que puedan ser utilizados en la práctica clínica de manera rutinaria. La identificación temprana de la enfermedad es crucial, ya que puede permitir un manejo más eficaz y potencialmente mejorar la calidad de vida del paciente a lo largo de la progresión de la enfermedad.

**¿Qué es la Enfermedad de Huntington?**
La enfermedad de Huntington (HD) es un trastorno neurodegenerativo autosómico dominante que afecta el sistema nervioso central y se caracteriza por una progresiva degeneración de neuronas, principalmente en áreas específicas del cerebro, como los ganglios basales. esta enfermedad se manifiesta de manera insidiosa, comenzando con cambios sutiles en el control motor, pero eventualmente avanzando hacia movimientos involuntarios (corea), deterioro cognitivo y una variedad de síntomas psiquiátricos, como depresión y ansiedad. Se estima que los síntomas motores, cognitivos y psiquiátricos pueden desarrollarse a lo largo de más de 30 años antes de que se realice un diagnóstico clínico formal, esto significa que la enfermedad puede estar presente en el cerebro mucho antes de que los pacientes sean conscientes de su condición, lo que complica tanto la identificación como la gestión de la enfermedad, además, existe una clara relación genética, ya que la presencia de una expansión del triplete CAG en el gen HTT es el causante directo de la enfermedad. Con el avance de la tecnología y la investigación, se han identificado diversos biomarcadores que podrían permitir una detección más temprana y precisa de la enfermedad, lo que representa una esperanza para el desarrollo de tratamientos más efectivos en el futuro.

**¿Cómo se Diagnostica?**
El diagnóstico de la enfermedad de Huntington tradicionalmente se basa en la evaluación clínica de los síntomas presentados por el paciente, con un enfoque particular en los signos motores que son característicos de la enfermedad, como los movimientos involuntarios y la dificultad para coordinar movimientos, sin embargo, dado que estos síntomas pueden no aparecer hasta más tarde en el curso de la enfermedad, se están considerando métodos adicionales para facilitar un diagnóstico más precoz. Las pruebas genéticas son una herramienta fundamental en este proceso, ya que permiten confirmar la presencia de la mutación en el gen HTT, la capacidad de identificar a los portadores de la mutación, incluso antes de que aparezcan los síntomas, abre la puerta a nuevas estrategias de intervención y a la posibilidad de realizar ensayos clínicos con poblaciones en riesgo, además, se están investigando marcadores biológicos y clínicos refinados, incluidos los hallazgos de neuroimágenes que pueden mostrar cambios en la morfología cerebral y en las funciones cognitivas mucho antes de que se cumplan los criterios diagnósticos tradicionales. A medida que se avanza en la investigación, se espera que estos métodos se integren en la práctica clínica, lo que podría transformar el enfoque hacia la identificación y el manejo de la enfermedad de Huntington.

**Retos para el Diagnóstico Temprano**
A pesar de los avances en la identificación de biomarcadores y en la comprensión de la enfermedad de Huntington, existen múltiples retos que dificultan el diagnóstico temprano, en primer lugar, hay una falta de conocimiento general sobre la enfermedad, tanto en el público como en algunos profesionales de la salud, lo que limita la capacidad de reconocer signos y síntomas tempranos, además, el diagnóstico de HD se ha vuelto un tema controvertido, con importantes implicaciones políticas, éticas y sociales. Esto se debe a la complejidad de la enfermedad y a las repercusiones que puede tener un diagnóstico prematuro en la vida del paciente y su familia, por otro lado, la validación de nuevos marcadores biológicos y clínicos sigue siendo una prioridad, ya que aunque se han identificado varios indicadores, estos aún necesitan ser comprobados en estudios más amplios antes de ser aceptados como criterios diagnósticos. Otro desafío es la lentitud en la inscripción de pacientes en ensayos clínicos, lo que retrasa el avance de la investigación y el desarrollo de tratamientos efectivos. Finalmente, el diagnóstico temprano puede conllevar efectos psicológicos adversos, como el aumento del estrés y la posibilidad de discriminación, lo que genera una resistencia natural a buscar pruebas diagnósticas antes de que los síntomas sean evidentes. Todos estos factores subrayan la necesidad de un enfoque más integral y colaborativo en la identificación y manejo de la enfermedad de Huntington, tanto en el ámbito clínico como en la investigación.

**¿Qué es la enfermedad de la esclerosis lateral amiotrófica (ALS)?**
La esclerosis lateral amiotrófica (ALS) es una enfermedad neurodegenerativa progresiva que afecta las neuronas motoras, las células nerviosas responsables del control de los músculos voluntarios en el cerebro y la médula espinal, esta enfermedad, a menudo conocida como enfermedad de Lou Gehrig, causa una pérdida gradual del control muscular, lo que resulta en debilidad, espasmos y atrofia muscular. A medida que la enfermedad avanza, los pacientes pueden experimentar dificultades para realizar actividades cotidianas como caminar, hablar y tragar, lo que afecta significativamente su calidad de vida, con el tiempo, la debilidad muscular se agrava, y puede comprometer los músculos responsables de la respiración, lo que lleva a la necesidad de dispositivos de asistencia respiratoria. Lamentablemente, no existe una cura para la ALS, y la mayoría de los pacientes fallecen dentro de un período de 14 a 18 meses después del diagnóstico, aunque algunos pueden vivir muchos años más. La naturaleza progresiva de la enfermedad y su impacto en múltiples aspectos de la vida del paciente la convierten en una de las afecciones más desafiantes y devastadoras dentro del ámbito de la neurología.

**Diagnóstico de la ALS**
El diagnóstico de la ALS es un proceso que requiere un enfoque meticuloso y multidisciplinario, inicialmente, un neurólogo realiza una evaluación clínica exhaustiva, que incluye una revisión completa de la historia médica del paciente y una evaluación de sus síntomas, los médicos buscan signos de debilidad muscular y atrofia, así como otros síntomas neurológicos que puedan indicar la presencia de esta enfermedad, además, se llevan a cabo exámenes neurológicos detallados para evaluar la función motora, lo que incluye pruebas de reflejos, fuerza y tono muscular. Para corroborar los hallazgos clínicos, se pueden solicitar estudios de imágenes, como resonancias magnéticas, que ayudan a descartar otras afecciones que presenten síntomas similares. Un componente crucial en el diagnóstico es la electromiografía (EMG), que mide la actividad eléctrica de los músculos y puede revelar daños en los nervios motores. En algunos casos, especialmente en aquellos con antecedentes familiares, se realizan pruebas genéticas para identificar formas hereditarias de la enfermedad. Sin embargo, el proceso puede ser complicado y puede requerir la exclusión de otras condiciones neurológicas antes de llegar a un diagnóstico definitivo de ALS.

**Retos para el diagnóstico temprano**
El diagnóstico temprano de la ALS enfrenta varios desafíos significativos que complican su identificación oportuna, uno de los mayores obstáculos es la presentación inicial de los síntomas, que a menudo son vagos y pueden ser fácilmente confundidos con otras afecciones menos graves, los primeros signos de ALS, como debilidad muscular, calambres y dificultades para realizar actividades diarias, pueden no ser suficientes para alertar a los médicos sobre la posibilidad de una enfermedad neurodegenerativa, además, actualmente no existen biomarcadores específicos que puedan confirmar el diagnóstico de ALS, lo que dificulta aún más su identificación en las etapas iniciales. La variabilidad en la presentación clínica de la enfermedad es otro factor que complica el diagnóstico; la ALS puede comenzar en diferentes partes del cuerpo y progresar a diferentes ritmos, lo que puede llevar a un diagnóstico erróneo. El acceso limitado a neurólogos especializados también puede resultar en demoras significativas en el diagnóstico, ya que no todos los pacientes tienen la posibilidad de consultar a un experto en trastornos del movimiento de inmediato. Finalmente, la falta de conocimiento y concienciación sobre la ALS entre los médicos y el público en general contribuye a diagnósticos erróneos o tardíos, lo que subraya la necesidad de educación y sensibilización en torno a esta enfermedad devastadora.

```python
def hampelfilter(wk, K, FilterParms):
    """
    Procedure to implement the Hampel filter
    """
    #
    Thresh = FilterParms[0]
    ctr = wk#[K]
    ref = np.median(wk)

    AbsDev = abs(wk - ref)
    MAD = 1.4826 * np.median(AbsDev)
    TestVal = abs(ctr - ref)
    #if TestVal > Thresh * MAD:
    yk = ctr
    numero = np.sum((TestVal > Thresh * MAD)==True)

    yk[TestVal > Thresh * MAD] = ref


    return yk
```

```python
# Esta es la función que calcula la entropía de permutación
def permutation_entropy(signal, dim=4, t=1):

  # Modulo numpy Función de ventana deslizante
  sorted_matrix = sliding_window_view(signal, dim, axis=0).T[:, ::t]
  # Patrones ordinales y ranking
  index_matrix = np.argsort(sorted_matrix, axis=0)
  #Busca lo arreglos diferentes y las veces que se repite
  _ , counts = np.unique(index_matrix, axis=1, return_counts=True)
  #numero de columnas de la matriz
  n = len(signal)- (dim-1) * t
  # frecuencia relativa de ocurrencia de cada permutación única en la matriz
  relative_freq = counts / n
  # Entropía de Permutación a partir de la entropia de shannon
  PE= np.sum(- relative_freq * np.log2(relative_freq))
  # Se retorna el valor de la entropia y la matriz de valores organizados
  return PE
```

# Pruebas Estadisticas

En el presente análisis, se inició con la evaluación de la normalidad de los datos mediante la prueba de Shapiro-Wilk, con el fin de determinar si las variables de interés seguían una distribución normal, este paso es crucial en el análisis estadístico, ya que la normalidad de los datos influye directamente en la selección de las pruebas adecuadas para la comparación entre grupos.

```python
import pandas as pd
import scipy.stats as stats

# Función para evaluar la normalidad de los datos
def evaluar_normalidad(data, variable):
    stat, p_value = stats.shapiro(data[variable])

    print(f"Resultados de la Prueba de Shapiro-Wilk para {variable}:")
    print(f"Estadístico: {format(stat, '.50f')}")
    print(f"Valor p: {format(p_value, '.50f')}")

    # Planteamiento de hipótesis
    if p_value > 0.05:
        print(f"Para la variable {variable}: No se rechaza la hipótesis nula (H₀). Los datos parecen seguir una distribución normal.")
        print("Prueba recomendada: Paramétrica\n")
        return "Paramétrica"
    else:
        print(f"Para la variable {variable}: Se rechaza la hipótesis nula (H₀). Los datos no siguen una distribución normal.")
        print("Prueba recomendada: No paramétrica\n")
        return "No paramétrica"

# Evaluar normalidad en todas las variables de interés
def evaluar_todas_las_variables(df, variables):
    resultados = {}
    for variable in variables:
        if variable in df.columns:
            tipo_prueba = evaluar_normalidad(df, variable)
            resultados[variable] = tipo_prueba
        else:
            print(f"La variable {variable} no se encuentra en el DataFrame.")

    return resultados

# Variables de interés
variables = ['sd_intervalo_izq', 'cv_intervalo_izq', 'ent_intervalo_izq',
             'sd_intervalo_der', 'cv_intervalo_der', 'ent_intervalo_der',
             'sd_balanceo_izq', 'cv_balanceo_izq', 'ent_balanceo_izq',
             'sd_balanceo_der', 'cv_balanceo_der', 'ent_balanceo_der']

# Aplicar la evaluación de normalidad a todas las variables
resultados_normalidad = evaluar_todas_las_variables(df_analisis, variables)

# Mostrar los resultados resumidos
#for variable, tipo_prueba in resultados_normalidad.items():
    #print(f"Variable: {variable} | Prueba recomendada: {tipo_prueba}")
```

En el contexto de este estudio, los resultados de la prueba de normalidad indicaron que la mayoría de las variables no cumplían con el supuesto de normalidad, lo que llevó a la decisión de descartar las pruebas paramétricas, que requieren esta condición.

A pesar de que un análisis adicional de homocedasticidad —es decir, la igualdad de varianzas entre grupos— es común en este tipo de análisis, se optó por no realizar esta evaluación. Esta decisión se fundamentó en que, dado que los datos ya mostraban una falta de normalidad, la evaluación de la homocedasticidad se volvía menos relevante en este contexto.

Las pruebas no paramétricas, como la prueba de Kruskal-Wallis,que es adecuada para comparar tres o más grupos independientes cuando los datos no cumplen con los supuestos de normalidad, no requieren que las varianzas sean homogéneas, lo que las convierte en una opción más apropiada para el análisis realizado.

```python
import pandas as pd
import scipy.stats as stats

# Función para realizar la prueba de Kruskal-Wallis
def kruskal_test(df, variable, group_column):
    grupos = [group[variable].values for name, group in df.groupby(group_column)]
    stat, p_value = stats.kruskal(*grupos)

    print(f"Resultados de la Prueba de Kruskal-Wallis para {variable}:")
    print(f"Estadístico: {format(stat, '.50f')}")
    print(f"Valor p: {format(p_value, '.50f')}")

    if p_value < 0.05:
        print("Se rechaza la hipótesis nula (H₀). Hay diferencias significativas entre los grupos.\n")
    else:
        print("No se rechaza la hipótesis nula (H₀). No hay diferencias significativas entre los grupos.\n")

# Variables de interés
variables = ['sd_intervalo_izq', 'cv_intervalo_izq', 'ent_intervalo_izq',
             'sd_intervalo_der', 'cv_intervalo_der', 'ent_intervalo_der',
             'sd_balanceo_izq', 'cv_balanceo_izq', 'ent_balanceo_izq',
             'sd_balanceo_der', 'cv_balanceo_der', 'ent_balanceo_der']

# Aplicar la prueba de Kruskal-Wallis para cada variable
for variable in variables:
    kruskal_test(df_analisis, variable, 'Estado')
```

Este enfoque se alinea con el objetivo de obtener resultados robustos y fiables, asegurando que las conclusiones sobre las diferencias entre los grupos clínicos (como Parkinson, ELA y Huntington) y el grupo control sean válidas, a pesar de la falta de normalidad en los datos. Así, se priorizó la validez de los resultados sobre la rigidez de los supuestos tradicionales en el análisis paramétrico. De este modo, se logró un análisis más acorde con las características de los datos disponibles, garantizando la integridad del estudio y la pertinencia de las inferencias realizadas.
