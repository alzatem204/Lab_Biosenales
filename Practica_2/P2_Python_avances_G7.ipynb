La funci√≥n rampa se define como:

2. Implemente una funci√≥n en Python que permita generar una funci√≥n rampa, definida en un
intervalo ùëõ1‚â§ùëõ0‚â§ùëõ2 con una pendiente m.

```python
def rampa (n0, n1, n2, m):

  n = np.arange(n1, n2 + 1)
  rampa = np.where(n < n0, 0, np.where(n <= n2, m * (n - n0), m * (n2 - n0)))

  return [rampa,n]
```

```python
n1 = 0
n0 = 2
n2 = 10
m = 0.5

rampa,n = rampa(n0, n1, n2, m)

plt.figure(figsize=(8, 5))
plt.stem(n, rampa, 'b')
plt.title('Funci√≥n Rampa')
plt.xlabel('n')
plt.ylabel('Valor de la Rampa')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.show()
```

```python
def impseq(n0, n1, n2):
# Genera x(n) = delta(n-n0); n1 <= n <= n2
# ----------------------------------------------
 n = np.arange(n1,n2+1) # Se crea el vector de muestras
 x = (n-n0) == 0
 return [x,n]
```

```python
n1 = -4
n2 = 1

# Crear los impulsos con las amplitudes correspondientes
imp1, n = impseq(-1, n1, n2)  # 3Œ¥(n + 1)
imp2, _ = impseq(-3, n1, n2)  # 5Œ¥(n + 3)
imp3, _ = impseq(-2, n1, n2)  # 3Œ¥(n + 2)
imp4, _ = impseq(0, n1, n2)   # 3Œ¥(n) + Œ¥(n)

# Crear la secuencia x1(n)
x1 = 3 * imp1 + 5 * imp2 + 3 * imp3 + 4 * imp4  # Sumar impulsos ponderados

# Graficar la funci√≥n compuesta por impulsos
plt.figure(figsize=(10, 6))
plt.stem(n, x1, basefmt=" ", use_line_collection=True)
plt.title('Funci√≥n compuesta de impulsos $x_1(n)$')
plt.xlabel('n')
plt.ylabel('$x_1(n)$')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```
