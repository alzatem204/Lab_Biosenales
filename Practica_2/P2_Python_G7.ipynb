**Rampa**

La funci√≥n rampa se define como:
ùëü(ùëõ ‚àí ùëõ0) = {
0 ùë†ùëñ ùëõ < ùëõ0
ùëõ ‚àí ùëõ0 ùë†ùëñ ùëõ ‚â• ùëõ0
2. Implemente una funci√≥n en Python que permita generar una funci√≥n rampa, definida en un
intervalo ùëõ1‚â§ùëõ0‚â§ùëõ2 con una pendiente m.

```python
def rampa (n0, n1, n2, m):

  n = np.arange(n1, n2 + 1)
  rampa = np.where(n < n0, 0, np.where(n <= n2, m * (n - n0), m * (n2 - n0)))

  return [rampa,n]
```

```python
n1 = 0
n0 = 2
n2 = 10
m = 0.5

rampa,n = rampa(n0, n1, n2, m)

plt.figure(figsize=(8, 5))
plt.stem(n, rampa, 'b')
plt.title('Funci√≥n Rampa')
plt.xlabel('n')
plt.ylabel('Valor de la Rampa')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```

```python
def impseq(n0, n1, n2):
# Genera x(n) = delta(n-n0); n1 <= n <= n2
# ----------------------------------------------
 n = np.arange(n1,n2+1) # Se crea el vector de muestras
 x = (n-n0) == 0
 return [x,n]
```

b. El siguiente sistema arroja una secuencia definida entre ‚àí6 ‚â§ ùëõ ‚â§ 6

```python
def unitario(n1, shift=0):
    n = np.arange(-6,7)# se inicializa en ese para poder graficar
    u = np.where(n >= n1, 1, 0)
    n -= shift
    return u, n

def main():
    u1,n1= unitario(0, shift=4)
    u2,n2= unitario(0, shift=2)

    v_1 = u1 * -2
    v_2 = u2 * 3

    resultado = v_1 + v_2
    return resultado, n1

if __name__ == "__main__":
    resultado, n = main()
    plt.figure(figsize=(8, 5))
    plt.stem(n,resultado,'b')
    plt.title('Secuencia 3b')
    plt.xlabel('Rango')
    plt.ylabel('Resultado')
    plt.grid(True)
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.show()
```

```python
def expo(n, n1, n2):
  e = np.exp(0.01 * n)
  return e

def seno(n, n1, n2):
  s = np.sin(2 * np.pi *n )
  return s

n=np.arange(0,101)
n1 = 0
n2 = n[-1]

exponencial = expo(n, n1, n2)
sinusoidal = seno(n,n1,n2)

secuencia = exponencial * sinusoidal

plt.figure(figsize=(8, 5))
plt.stem(n,secuencia,'b')
plt.title('Secuencia 3c')
plt.xlabel('Rango')
plt.ylabel('Resultado')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid()
plt.show()
```

4. Genere la siguiente secuencia

ùë•[ùëõ] = {
ùëü[ùëõ] 0 ‚â§ ùëõ ‚â§ 5

ùëü[ùëõ ‚àí 5] 6 ‚â§ ùëõ ‚â§ 11

ùëü[ùëõ ‚àí 10] 12 ‚â§ ùëõ ‚â§ 17}

```python
def rampa (n0, n1, n2, m):

  n = np.arange(n1, n2 + 1)
  rampa = np.where(n < n0, 0, np.where(n <= n2, m * (n - n0), m * (n2 - n0)))

  return [rampa,n]

rampa1, n1 = rampa(0, 0, 5, 1)

# Segundo tramo: r[n-5] para 6 ‚â§ n ‚â§ 11
rampa2, n2 = rampa(5, 6, 11, 1)

# Tercer tramo: r[n-10] para 12 ‚â§ n ‚â§ 17
rampa3, n3 = rampa(10, 12, 17, 1)

# Concatenar las rampas y los valores de n
x_con = np.concatenate((rampa1, rampa2, rampa3))
n_con = np.concatenate((n1, n2, n3))
```

```python
plt.figure(figsize=(8, 5))
plt.stem(n_con, x_con, 'b')
plt.title('Funci√≥n Rampa')
plt.xlabel('n')
plt.ylabel('Valor de la Rampa')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.show()
```
